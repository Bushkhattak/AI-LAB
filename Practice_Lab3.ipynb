```python
my_string = "Python Programming"
# Get every second character in the string
print(my_string[::2])
# Reverse the string
print(my_string[::-1])
```

```python
#List slicing in Python
my_list = ['p', 'r', 'o', 'g', 'r', 'a' , 'm', 'i' , 'z']
#items from index 2 to index 4
print(my_list[2:4])
#items from index 5 to end
print(my_list[5:])
#items beginning to end
print(my_list[:])
```

```python
#using append
numbers = [1, 2, 32, 56]
numbers.append(45)
print(numbers)
```

```python
#using extend
my_list = [1, 2, 3]
another_list = [4, 5, 6]
my_list.extend(another_list)
print(my_list)
```

```python
#Change List Items
my_list = ["Python" , "Java" , "C++"]
#changing the third item to C
my_list[2] = 'C'
print(my_list)
```

```python
#Remove an item from List
my_list = ["Python" , "Java" ,"C++"]
del my_list[1]
print(my_list)
#by using remove
list = ["Python" , "Java"]
list.remove("Python")
print(list)
```

```python
#Check if an item Exists in the Python List
my_list = ["Python" , "Java" , "C++"]
print('C' in my_list)
print('C++' in my_list)
print('Java' in my_list)
```

```python
#List Length
list = ["Python" , "Java" ,  "C++" , "C"]
print(len(list))
```

```python
#Python List Comprehension
new_var = [number*number for number in range(1,6)]
print(new_var)
```

```python
#Loop
my_list = ["banana" , "apple" , "strawberry"]
for x in my_list:
 print(x)
```

```python
#Nested
my_list = ["banana" , "apple" , "strawberry"]
cur = ['Python' , 'Java', 'C++']
for x in my_list:
  for y in cur:
    print(x , y)
```
